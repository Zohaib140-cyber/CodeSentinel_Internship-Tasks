```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = {
    "Product": ["Shoes", "Bags", "Shirts", "Shoes", "Bags", "Shirts", "Shoes", "Bags", "Shirts"],
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West", "North"],
    "Sales": [250, 180, 200, 300, 220, 190, 270, 210, 230],
    "Profit": [50, 40, 30, 70, 60, 35, 55, 45, 50]
}

df = pd.DataFrame(data)

print("Dataset Preview:")
print(df.head())

# --- 1. Bar Plot (Sales by Product)
plt.figure(figsize=(6,4))
sns.barplot(x="Product", y="Sales", data=df, estimator=sum)
plt.title("Total Sales by Product")
plt.show()

# --- 2. Histogram (Sales Distribution)
plt.figure(figsize=(6,4))
plt.hist(df["Sales"], bins=5, edgecolor="black")
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()

# --- 3. Pie Chart (Sales by Region)
region_sales = df.groupby("Region")["Sales"].sum()
plt.figure(figsize=(6,6))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution by Region")
plt.show()

# --- 4. Boxplot (Profit by Product)
plt.figure(figsize=(6,4))
sns.boxplot(x="Product", y="Profit", data=df)
plt.title("Profit Spread by Product")
plt.show()
```
